```python
#Pore throat Using Numpy
k_mD_02 = np.arange(0.001, 0.5, 0.001)
k_mD_05 = np.arange(0.01, 3, 0.01)
k_mD_1 = np.arange(0.05, 12, 0.05)
k_mD_2 = np.arange(0.2, 60, 0.2)
k_mD_4 = np.arange(0.5, 260, 0.5)
k_mD_10 = np.arange(3, 2100, 3)
k_mD_20 = np.arange(10, 10000, 10)
k_mD_40 = np.arange(30, 50000, 30)
```

```python
#Rp35
rp35_02 = 0.2
rp35_05 = 0.5
rp35_1 = 1
rp35_2 = 2
rp35_4 = 4
rp35_10 = 10
rp35_20 = 20
rp35_40 = 40
```

```python
def phi_rp35 (K, rp35):
    phi_rp35 = 1/(((rp35/2.665)**(1/0.45))/K)
    return phi_rp35
```

```python
df_02 = pd.DataFrame(data = k_mD_02, columns = ['K_mD'])
df_02['rp35'] = rp35_02
df_02['Porosity'] = phi_rp35 (df_02['K_mD'], df_02['rp35'])
df_02
```
